```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
header = ["age", "Job-type", "fnlwgt", "edu", "edu-num", "marital-status", "Job", "Relationship", "race", "sex", "gain", "loss", "hrs/week", "naive", "income"]
df = pd.read_csv('adult/adult.data', low_memory = False, names=header)
df.head()
```

```python
hrs_data = df.groupby(["hrs/week", "income"]).size().unstack(fill_value=0)
ax = hrs_data.plot(kind="bar", stacked=True, figsize=(14, 8))
plt.xlabel("Worked Hours per Week")
plt.ylabel("Total Count per Hour")
plt.title("Weekly Worked Hours Income")
ax.legend(title="Income Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
